```python
import yfinance as yf

# 获取 5238.KL 股票的财务数据
stock = yf.Ticker("5238.KL")

# 获取财务报表
financials = stock.financials
balance_sheet = stock.balance_sheet
cashflow = stock.cashflow

# 打印财务数据
print("财务报表：")
print(financials)

print("\n资产负债表：")
print(balance_sheet)

print("\n现金流量表：")
print(cashflow)

financials.to_excel("financials.xlsx")
balance_sheet.to_excel("balance_sheet.xlsx")
cashflow.to_excel("cashflow.xlsx")
```
